Question 1: If a positive integer is entered through the keyboard, write a recursive function to obtain the prime factors of the number.

```python
def primefac(n):
  if n == 1:
    return []
  for i in range(2, n+1):
    if n % i == 0:
      return [i] + primefac(n // i)
n=int(input("Enter a number: "))
print(primefac(n))
```

Question 2:	A positive integer is entered through the keyboard. Write a function to find its binary equivalent of this number.

```python
def bin(n):
  if n:
    return n%2 + (10*bin(n//2))
  else:
    return n
bin(int(input("Enter a no. :")))
```

Question 3: A string is entered through the keyboard. Write a recursive function that counts the number of vowels in this string.

```python
def str_count(string,index=0,count=0):
  if index == len(string):
    return count
  if string[index].lower() in 'aeiou':
    count += 1
  return str_count(string, index + 1, count)
input_string = input("Enter a string: ")
vowel_count = str_count(input_string)
print("Number of vowels:", vowel_count)
```

Question 4: Write a recursive function that reverses the list of numbers that it receives.

```python
def revlist(lst):
  if not lst:
    return []
  return [lst[-1]] + revlist(lst[:-1])
a=(input("Enter ele separated by space:")).split(" ")
revlist(a)
```

Question 4: Calculate a^b where a and b received through the keyword using recursion.

```python
def power(a, b):
  if b == 0:
    return 1
  elif b == 1:
    return a
  else:
    return a * power(a, b - 1)
a = int(input("Enter the base (a): "))
b = int(input("Enter the exponent (b): "))
print(power(a, b))
```

Question 5: A list contains some negative and some positive values. Write a recursive function that sanitizes the list by replacing all negative numbers with 0.

```python
def nonegative_list(lst):
  if not lst:
    return []
  if lst[0] < 0:
    return [0] + nonegative_list(lst[1:])
  else:
    return [lst[0]] + nonegative_list(lst[1:])

mylist=input("Enter ele separated by space:").split(" ")
mylist=[int(i) for i in mylist]
print(nonegative_list(mylist))
```
